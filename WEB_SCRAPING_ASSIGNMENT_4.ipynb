# Q1

Scrape the details of most viewed videos on YouTube from Wikipedia.

Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos

You need to find following details:

A) Rank

B) Name

C) Artist

D) Upload date

E) Views

```python
# connect to webdriver
driver=webdriver.Chrome(r"chromedriver.exe")
driver.maximize_window()

# search on google
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")

#fetch Require Data
rank=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
Rank=[i.text for i in rank]

name=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')
Name=[i.text for i in name]

artist=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')
Artist=[i.text for i in artist]

upload_date=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
Upload_date=[i.text for i in upload_date]

views=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
Views=[i.text for i in views]

#Creat Data Frame
df=pd.DataFrame({
    'Rank':Rank[0:30],
    'Name':Name[0:30],
    'Artist':Artist[0:30],
    'Upload_date':Upload_date[0:30],
    'Views':Views[0:30]
})
df
```

```python
# connect to webdriver
driver=webdriver.Chrome(r"chromedriver.exe")
driver.maximize_window()

# search on google
driver.get("https://archive.ics.uci.edu/ml/index.php")


# for click
Click=driver.find_element(By.XPATH,'/html/body/table[2]/tbody/tr/td/span/b/a')
Click.click()
time.sleep(3)


#fetch all require Data
dataset_name=driver.find_elements(By.XPATH,'//table[@cellpadding="3"]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]/table/tbody/tr/td[2]/p/b')
Dataset_name=[i.text for i in dataset_name]

data_type=driver.find_elements(By.XPATH,'//table[@cellpadding="3"]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]/p')
Data_type=[i.text for i in data_type]


task=driver.find_elements(By.XPATH,'//table[@cellpadding="3"]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]/p')
Task=[i.text for i in task]

attribute_type=driver.find_elements(By.XPATH,'//table[@cellpadding="3"]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]/p')
Attribute_type=[i.text for i in attribute_type]

no_of_instances=driver.find_elements(By.XPATH,'//table[@cellpadding="3"]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]/p')
No_of_instances=[i.text for i in no_of_instances]

no_of_attribute=driver.find_elements(By.XPATH,'//table[@cellpadding="3"]/tbody/tr/td[2]/table[2]/tbody/tr/td[6]/p')
No_of_attribute=[i.text for i in no_of_attribute]

year=driver.find_elements(By.XPATH,'//table[@cellpadding="3"]/tbody/tr/td[2]/table[2]/tbody/tr/td[7]/p')
Year=[i.text for i in year]


#Creat DataFrame
df=pd.DataFrame({
    'Dataset_name':Dataset_name,
    'Data_type':Data_type[1:],
    'Task':Task[1:],
    'Attribute_type':Attribute_type[1:],
    'No_of_instances':No_of_instances[1:],
    'No_of_attribute':No_of_attribute[1:],
    'Year':Year[1:]
})
df
```
